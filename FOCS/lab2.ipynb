### First approach - linear search $O(n)$
We start with some random initialization, don't worry about the technicalities, we will cover them later. Just know, that N and T are created according to the problem size n. What is the complexity of the initialization?


#### Initialization

```python
import string, random
n = 300000
NAME = "ALIS"
N = [''.join(random.choices(string.ascii_uppercase, k=4)) for i in range(n)]
T = [random.randint(971500000000, 971569999999) for i in range(n)]
```

```python
N[0]
```

```python
%%timeit
## your code here
## do not include print-out in the timer, rather, create a result variable
i = 0
FOUND = False
while not FOUND and i < n:
	if NAME == N[i]:
		result = T[i]
		FOUND = True
	i += 1
if not FOUND:
	result = "Sorry, this name is not in the directory"
```

```python
## do not include print-out in the timer
print(result)
```

### Second approach - dictionaries
Remember from the lecture, a very quick way to get information is through the convenient dictionary data structure. We first create one 

The dictionary lookup is $O(1)$ (on average), which should yield faster results than the linear search

```python
## creating a dictionary from two parallel results
phonebook = dict(zip(N, T))

	
```

```python
%%timeit
if NAME in phonebook:
	result = phonebook[NAME]
else:
	result = "not found"
```

```python
%%timeit
i = 0
```
